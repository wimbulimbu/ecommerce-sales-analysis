```python
orders_table = pd.read_sql('''
SELECT *
FROM orders                           
                           ''', conn, parse_dates=['created_at'])
```

```python
orders_table['profit'] = orders_table['price_usd'] - orders_table['cogs_usd']
orders_table['month'] = orders_table['created_at'].dt.to_period('M')
```

```python
total_revenue = orders_table['price_usd'].sum()
total_profit = orders_table['profit'].sum()
total_buying_customers = orders_table['user_id'].nunique()
total_items_sold = orders_table['items_purchased'].sum()
date_range = f'{orders_table['created_at'].min()} - {orders_table['created_at'].max()}'
profit_margin = (total_profit/total_revenue) * 100
average_order_value = total_revenue/len(orders_table)
```

```python
orders_table_monthly = orders_table.groupby(['month'], as_index=False).agg(
    total_revenue = ('price_usd', 'sum'),
    total_profit = ('profit', 'sum'),
    total_orders = ('items_purchased', 'sum')
)
orders_table_monthly['month'] = orders_table_monthly['month'].astype(str)
```

```python
fig = px.line(orders_table_monthly, x='month', y='total_revenue', title='revenue_by_month')
fig.show()
```

```python
product_data_sum = pd.read_sql('''                           
SELECT 
    p.product_name, 
    COUNT(oi.order_item_id) as total_orders,
    SUM(oi.price_usd) as total_revenue,
    SUM(oi.price_usd - oi.cogs_usd) as total_profit,
    ROUND(SUM(oi.price_usd - oi.cogs_usd) * 100.0/ SUM(oi.price_usd), 2) as profit_margin
                                                       
FROM order_items oi
LEFT JOIN products p
ON oi.product_id = p.product_id       
GROUP BY p.product_name
ORDER BY total_profit DESC                
                           ''', conn)
```

```python
product_performance = pd.read_sql('''
SELECT 
    p.product_name, 
    oi.created_at, 
    oi.price_usd,
    oi.cogs_usd,
    oi.product_id,
    oi.order_item_id                                 
FROM order_items oi
LEFT JOIN products p
ON oi.product_id = p.product_id                                  
                                  ''', conn, parse_dates=['created_at'])
```

```python
product_performance['monthly'] = product_performance['created_at'].dt.to_period('M')
product_performance['profit'] = product_performance['price_usd'] - product_performance['cogs_usd']
```

```python
product_performance_data = product_performance.groupby(['monthly', 'product_name'], as_index=False).agg(
    total_orders = ('order_item_id', 'count'),
    total_revenue = ('price_usd', 'sum'),    
).sort_values(by='monthly')
product_performance_data['monthly'] = product_performance_data['monthly'].astype(str)
```

```python
product_refunds = pd.read_sql('''
SELECT 
    p.product_name,
    SUM(oi.price_usd) as total_product_revenue,
    COALESCE(SUM(oir.refund_amount_usd), 0) as total_refund_amount,
    COUNT(oir.order_item_refund_id) as total_product_refunds,
    COUNT(oi.order_item_id) as total_product_sold
    
    
FROM order_items oi
LEFT JOIN order_item_refunds oir
ON oi.order_item_id = oir.order_item_id
INNER JOIN products p
ON oi.product_id = p.product_id
GROUP BY p.product_name
ORDER BY total_refund_amount DESC                                                             
                              ''', conn)
```

```python
product_refunds['product_refund_rate'] = ((product_refunds['total_product_refunds']/product_refunds['total_product_sold']) * 100).round(2)
product_refunds.sort_values(by='product_refund_rate', ascending=False)
```

```python
'''
Product Refund Insights

The Birthday Sugar Panda had the highest refund rate out of all the products. In terms of the total refund amount, The Original Mr. Fuzzy is the 
highest likely due to it being on the market the longest and having the most units sold. Yet the refund rate is lower than The Birthday Sugar Panda, which
in terms of the how long it has been on the market and the number of products sold. which is similar to The Forever Love Bear and The Hudson River Mini Bear.
The Birthday Sugar Panda has a significantly higher refund rate.

'''
```

```python
marketing_info = pd.read_sql('''
SELECT 
    ws.website_session_id,
    o.order_id,
    o.website_session_id,
    ws.device_type,
    ws.utm_source,
    ws.utm_campaign
FROM website_session ws
LEFT JOIN orders o
ON ws.website_session_id	= o.website_session_id                         
                             ''', conn)
```

```python
marketing_info
```

```python
device_type_info = marketing_info.groupby(['device_type'], as_index=False).agg(
    sessions = ('website_session_id', 'count'),
    orders = ('order_id', 'count')
)
device_type_info['conversion_rate'] = (device_type_info['orders']/ device_type_info['sessions'] * 100).round(2)
```

```python
device_type_info
```

```python
utm_source_info = marketing_info.groupby(['utm_source'], as_index=False).agg(
    sessions = ('website_session_id', 'count'),
    orders = ('order_id', 'count')
)
utm_source_info['conversion_rate'] = (utm_source_info['orders']/ utm_source_info['sessions'] * 100).round(2)
```

```python
utm_source_info
```

```python
utm_campaign_info = marketing_info.groupby(['utm_campaign'], as_index=False).agg(
    sessions = ('website_session_id', 'count'),
    orders = ('order_id', 'count')
)
utm_campaign_info['conversion_rate'] = (utm_campaign_info['orders']/ utm_campaign_info['sessions'] * 100).round(2)
```
